```python
# Import libraries for model testing
import pandas as pd
import pickle
import tensorflow as tf
from sklearn.metrics import classification_report
```

```python
# Load model
model = tf.keras.models.load_model(
"/content/drive/MyDrive/Fraud Detection/models/fraud_nn_smote_v1.h5"
)
```

```python
# Load scaler
with open("/content/drive/MyDrive/Fraud Detection/models/scaler_smote_v1.pkl","rb") as f:
    scaler = pickle.load(f)
```

```python
# Load new transactions
df = pd.read_csv("/content/drive/MyDrive/Fraud_Detection_Project/creditcard_2023.csv")
df.head()
```

```python
X = df.drop(['Class', 'id'], axis=1)
y = df['Class']
```

```python
df.head()
```

```python
# Scale
X_scaled = scaler.transform(X)
```

```python
# Predict
y_pred_prob = model.predict(X_scaled)
y_pred = (y_pred_prob > 0.5).astype(int)

print(classification_report(y, y_pred))
```
